### BTVN1: Cho ma trận  
```
= [
    [99,99,99],
    [99,99,99],
    [99,99,99]
]
```

+ hãy tưởng tượng ô 99 trống
+ giả sử 1 là x
+ giả sử 0 là o

+ lấy đầu vào từ phía X và phía o luân phiên
+ nếu đầu vào từ phía X là (0,0) thì ma trận sẽ trở thành
[[99,99,99],
[99,99,99],
[99,99,99]]
+ nếu đầu vào từ phía o cũng là (0,0) thì thông báo nhập lại, nếu không thì điền vào ma trận
+ thử thách, nếu người đầu tiên có 3 phần tử thì dừng trò chơi.

````python
import numpy as np

# Khởi tạo ma trận 3x3 với giá trị 99 (trống)
board = np.full((3, 3), 99, dtype=int)

# Định nghĩa người chơi
PLAYER_X = 1
PLAYER_O = 0
EMPTY = 99

# Hàm để in bảng
def print_board(board):
    symbols = {PLAYER_X: 'X', PLAYER_O: 'O', EMPTY: ' '}
    for row in board:
        print("| " + " | ".join(symbols[cell] for cell in row) + " |")
        print("-----------")

# Hàm kiểm tra chiến thắng
def check_win(board, player):
    # Kiểm tra hàng ngang
    for row in board:
        if np.all(row == player):
            return True
    # Kiểm tra cột dọc
    for col in board.T:
        if np.all(col == player):
            return True
    # Kiểm tra đường chéo chính
    if np.all(np.diag(board) == player):
        return True
    # Kiểm tra đường chéo phụ
    if np.all(np.diag(np.fliplr(board)) == player):
        return True
    return False

# Hàm kiểm tra hòa
def check_draw(board):
    return EMPTY not in board

# Bắt đầu trò chơi
current_player = PLAYER_X
print("Bắt đầu trò chơi Cờ caro (Tic-Tac-Toe)!")
print_board(board)

while True:
    try:
        player_symbol = 'X' if current_player == PLAYER_X else 'O'
        print(f"Lượt của người chơi {player_symbol}:")
        row = int(input("Nhập hàng (0, 1, hoặc 2): "))
        col = int(input("Nhập cột (0, 1, hoặc 2): "))

        # Kiểm tra tính hợp lệ của nước đi
        if not (0 <= row <= 2 and 0 <= col <= 2):
            print("Nước đi không hợp lệ. Hàng và cột phải nằm trong khoảng từ 0 đến 2.")
            continue
        if board[row, col] != EMPTY:
            print("Ô này đã có người chiếm. Vui lòng chọn ô khác.")
            continue

        # Cập nhật bảng
        board[row, col] = current_player
        print_board(board)

        # Kiểm tra chiến thắng
        if check_win(board, current_player):
            print(f"Chúc mừng người chơi {player_symbol} đã thắng cuộc!")
            break

        # Kiểm tra hòa
        if check_draw(board):
            print("Trò chơi hòa!")
            break

        # Chuyển lượt chơi
        current_player = PLAYER_O if current_player == PLAYER_X else PLAYER_X

    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập số nguyên.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
````

````python
import numpy as np
from sklearn.model_selection import train_test_split

# 1. Tạo một mảng numpy có kích thước 150 x 5 với dữ liệu ngẫu nhiên
# Sử dụng np.random.rand để tạo dữ liệu ngẫu nhiên giữa 0 và 1, sau đó nhân với một số để có phạm vi dữ liệu hợp lý (ví dụ: chiều cao, cân nặng, v.v.)
data_array = np.random.rand(150, 5)
print("Mảng dữ liệu ban đầu (150x5):")
print(data_array[:5]) # In ra 5 hàng đầu để xem trước
````

````python
# 2. Tách mảng dựa trên 4 cột đầu tiên thành X và cột cuối cùng thành y
X = data_array[:, :4]
y = data_array[:, 4]

print("\nKích thước của X (các đặc trưng):", X.shape)
print(X[:5])
print("\nKích thước của y (nhãn/mục tiêu):", y.shape)
print(y[:5])
````

````python
# 3. Tách X thành X_train và X_test (70% dữ liệu cho X_train)
# 4. Tách y thành y_train và y_test (70% dữ liệu cho y_train)
# Sử dụng train_test_split từ sklearn để thực hiện việc này

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("\nKích thước của X_train:", X_train.shape)
print("Kích thước của X_test:", X_test.shape)
print("Kích thước của y_train:", y_train.shape)
print("Kích thước của y_test:", y_test.shape)
````
