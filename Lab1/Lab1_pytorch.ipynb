### BTVN2 : tạo 1 tensor ban đầu có giá trị là 2. định nghĩa hàm số và tính gradient 
y = x^3 + 2x^2 + 5x + 1. hãy tính dy/dx tại giá trị của x 
dùng phương pháp gradient descent với: learning_rate alpha =0.1 để cập nhật giá trị x trong 10 vòng lặp


```python
import torch
x = torch.tensor(2.0, requires_grad=True)
learning_rate_alpha = 0.1
for i in range(10):
    y = x**3 + 2*x**2 + 5*x + 1 
    y.backward()
    with torch.no_grad():
        x -= learning_rate_alpha * x.grad  #cập nhật x theo công thức gradient descent
        x.grad.zero_() # reset gradient để tránh cộng dồn
    print(f'Iter {i+1} : x = {x.item()}')
```

### BTVN3: tạo 1 tập dữ liệu giả lập với x là số giờ học ngẫu nhiên từ 1 đến 10 và y là số điểm đc tính theo công thức y = 3x + 5+ noise , với noise là 1 giá trị ngẫu nhiên nhỏ 
1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad= true
2. tính MSE
3. Tính gradient
4. Cập nhật tham số w và b bằng gradient Descent với Learning rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp và quan sát sự hội tụ của mô hình

```python
import torch
x = torch.rand(10,1) * 9 + 1  #tạo 10 gtri x ngẫu nhiên tù 1 đến 10
noise = torch.randn(1) # tạo 1 gtri noise ngẫu nhiên nhỏ
y =  3*x + 5 + noise

# 1. khởi tạo tham số w và b ngẫu nhiên 
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print(f'tập gtri x: ',"\n", f'{x}')
print(f'noise : {noise}')
print(f'y = {y}')
print(f'w : {w.item()}')
print(f'b : {b.item()}')
```

```python
#2. tính MSE
y_pred = w * x + b  # dự đoán theo mô hình tuyến tính
loss = ((y_pred - y) ** 2).mean()  # cthuc mse = trung bình **2 sai số

print("MSE : ", loss.item())
```

```python
#3. Tính gradient
loss.backward()  #tính grad của loss theo w và b
print(f'gradient w : {w.grad.item()}')
print(f'gradient b : {b.grad.item()}')
```

```python
print(w.grad)
```

```python
#4. Cập nhật tham số w và b bằng gradient Descent với Learning rate alpha = 0.01
lr_alpha = 0.01
with torch.no_grad():
    w -= lr_alpha * w.grad
    b -= lr_alpha * b.grad

    w.grad.zero_()
    b.grad.zero_()
print(f'w cập nhật: {w.item()} ')
print(f'b cập nhật: {b.item()} ')
```

```python
#5. Lặp lại quá trình trên trong 100 vòng lặp và quan sát sự hội tụ của mô hình
for epoch in range(100):
    y_pred = w * x + b  # dự đoán theo mô hình tuyến tính
    loss = ((y_pred - y) ** 2).mean()  # cthuc mse = trung bình **2 sai số
    loss.backward() 

    lr_alpha = 0.01
    with torch.no_grad():
        w -= lr_alpha * w.grad
        b -= lr_alpha * b.grad

        w.grad.zero_()
        b.grad.zero_()
    print(f"Epoch {epoch}: Loss = {loss.item()}")
```
